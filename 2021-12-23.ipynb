```python
import numpy as np
import sympy as sp
import math
from sympy.plotting import plot
from sympy import I
from math import exp

sp.init_printing(use_latex=True)
```

```python
def Potential_d(x1,y1,z1):
    d_coord=np.array([x1,y1,z1])
    return d_coord

def Potential_epsilon(x2,y2,z2):
                     
    epsil_coord=np.array([x2,y2,z2])
    epsil_size=np.linalg.norm(epsil_coord)
    epsil_unit=epsil_coord/epsil_size
    
    return epsil_unit

def Potential_V(phi,E):
    Var1=np.array([0,0,0,0,0,0])
                     
    for i in range(0,6,1):
        Var1[i]=float(input())
        
    Potential_d(Var1[0],Var1[1],Var1[2])
    Potential_epsilon(Var1[3],Var1[4],Var1[5])
                     
    Ex_E_field=Potential_epsilon*2*E
    Potential=Potential_d*Ex_E_field*exp**(I*phi)
    
    return Potential
```

```python
t=sp.symbols('t')
a1=sp.symbols('a1',cls=sp.Function)
b1=(a1(t))
a2=sp.symbols('a2',cls=sp.Function)
b2=(a2(t))

phi=float(input("Value Phi"))
E=float(input("Value E"))

hbar=1.05457162000000*10**34
V=Potential_V(phi,E)
```
