```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import math 
```

```python
theta=np.linspace(0,2*np.pi,50)
def z_cord():
    a=4*(np.cos(theta)**2+1)
    return a

r=z_cord()/(np.cos(theta)**2+1)
```

```python
fig=plt.figure()
ax=fig.add_subplot(polar=True)

ax.plot(theta,r)
plt.show()
```
