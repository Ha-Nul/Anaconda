```python
import sympy as sp
```

```python
sp.init_printing(use_latex=True)
```

```python
t=sp.symbols('t')
a=sp.symbols('a',cls=sp.Function)
y=a(t)
```

```python
d_eq=sp.Eq(t*sp.diff(y,t)+y,2*t)
```

```python
sp.dsolve(d_eq,y)
```
